## El enfoque de Halmos
A diferencia de Lipschutz, Halmos se centra en la **teoria axiomatica**. No nos interesa solo operar conjuntos, si no definir bajo que reglas un conjunto es valido.


## 1. El Axioma de Extension
*"Dos conjuntos son iguales si y solo si tienen los mismos elementos."*

Este axioma implica que la **naturaleza** de un conjunto esta determinada unicamente por sus miembros, no por su orden o por la forma en que se describen.

```python
# --- EXPERIEMTO DE HALMOS: AXIOMA DE EXTENSION ---

# 1. Definicion por enumeracion (Extension)
A = {1, 2, 3,}

# 2. Definicion por propiedad (Comprension)
B = {x for x in range(1, 4)}

# 3. Definicion con desorden y duplicados (Capa de abstraccion)
C = {3, 1, 2, 2, 3}
# Verificacion de igualdad segun el Axioma de Extension
igual_AB = A == B
igual_BC = B == C

# --- RESULTADOS ---
print(f"conjuto A: {A}")
print(f"conjuto B: {B}")
print(f"conjuto C: {C}")
print(f"¿A es igual a B? (Diferente definicion): {igual_AB}")
print(f"¿B es igual a C? (Orden y duplicados): {igual_BC}")
```

## 2. El Axioma de Especificacion 
Halmos explica que no podemos crear conjuntos de la nada (para evitar paradojas). Necesitamos un conjunto existente $A$ y una condicion $P(x)$ para formar un nuevo conjunto:
$$B = \{x \in A \mid P(x)\}$$

En programacion, esto es lo que conocemos como **Filtrado de Datos**.

```python
# --- EXPERIMENTO DE ESPECIFICACION ---

# Tenemos un conjunto "Universo" inical
A = {x for x in range(1, 21)} # Numero del 1 al 20

# Especificamo una condicion : "x es un numero primo"
# (Usamos una funcion pequeña para la propiedad P(x))
def es_primo(n):
    if n <= 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

# Aplicamos el Axioma de especificacion
B = {x for x in A if es_primo(x)}

print(f"Conjunto base A: {A}")
print(f"Conjunto especificado B (Pirmos): {B}")
print(f"¿Es B subconjunto de A?: {B.issubset(A)}")
```

## 3 El Axioma del Par
Halmos Postula: *"Dados dos objetos (Pueden ser conjuntos), existe un conjunto que los contiene a ambos y a nadie mas"*

Este axioma es el que crear estructura. Si tengo $a$ y $b$, puedo crear $\{a, b\}$. si $a = b$, entonces creamos el **singulate** $\{a\}$.

```python
# --- EXPERIMENTO DEL PAR ---

a = 10
b = 20

# Creamos el para {a, b}
par = {a, b}

# Creamos un singulate (a = b)
singulate = {a, a}

print(f"Objeto a: {a}, Objeto b: {b}")
print(f"Conjunto par: {par}")
print(f"Conjunto Singulate (notar que la repiticion desaparece): {singulate }")
```

## 4. El Axioma de la Union (Colecciones)
Para Halmos, la union no es lolo juntar dos conjuntos. Es tomar una **familia** de conjuntos y extraer todos sus elemntos para formar un conjunto nuevo "mas grande"
Si tenemos $\matchcal{C} = \{ \{1, 2\}, \{3, 4\}, \{5\} \}$, la union de la coleccion es $\{1, 2, 3, 4, 5\}$.

```python
# --- EXPERIMENTO DE UNION DE COLLECCIONES ---

# Una colecion  (familia) de conjuntos
familia = [{1, 2}, {3, 4}, {4, 5}, {6}]

# El Axioma de la union: Extraer tolos los elementos
union_halmos = set()
for conjunto in familia:
    union_halmos.update(conjunto)

print(f"Familia de conjuntos: {familia}")
print(f"Unión de los conjuntos (Axioma de la unión): {union_halmos}")
```

## 11 (Codigo)
Halmos dice: *"Para cada conjunto existe una coleccion de conjuntos que contiene todos los subconjuntos del conjunto dado"*.

Si Lipschutz nos enseño *como* calcularlo, Halmos nos dice que la **esistencia** de este conjunto esta grantizada por las leyes del universo matematico.

```python
# --- EXPERIMENTO: EL CONJUNTO DE LAS PARTES ---

def axioma_potencia(S):
    elementos = list(S)
    n = len(elementos)
    return[set(elementos[j] for j in range(n) if (i & (1 << j))) for i in range(2**n)]

# Probemos con un conjunto de 3 elementos
A = {'Alpha', 'Beta', 'Gamma'}
P_A = axioma_potencia(A)

print(f"Conjunto A: {A}")
print(f"¿Cuantos subconjuntos existen? (2^3): {len(P_A)}")
print(f"Conjunto de potencias P(A): {P_A}")
```
