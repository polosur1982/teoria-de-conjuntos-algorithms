### 1.1 Defenicion
Un conjunto es una coleccion de objetos bien definidos. los objetos se llaman **elementos** o miembros.
- Si $a$ es un elemento del conjunto $A$, escribimos $a \ in A$
- Si  $a$ no pertenece a $A$ escribimos $a \ notin A$.

### 1.2 Especificacion del conjunto
1. **Metodo de Extension (Tabulacion):** Listar elementos entre llave. Ej: $A = \{1, 2, 3\}$.
2. **Metodo de Comprension (Abstraccion):** Definir mediante una propiedad. Ej: $B = \{X \mid x \text{ es par}\}$.

```python
# --- REGLA DE ORO DEL MIT: los set en python no admiten duplicados ---

# 1. Definicion por extension
# Lipschutz: "Conjuntos de vocales"
vocales = {'a', 'e', 'i', 'o', 'u'}

# 2. Definicion por comprension (Set comprehension)
# Lipschutz: "Conjuntos S de numeros enteros tales que x^2 = 9"
# S =  {x | x ∈ Z, x² = 9}
S = {x for x in range(-10, 11) if x**2 == 9}

# 3. Verificacion de pertenencia  ( Operador 'in')
pertenece_3 = 3 in S   # True
pertenece_5 = 5 not in S   #True

#--- SALIDA DE DATOS ---
print(f"Vocales (Extension): {vocales}")
print(f"El 3 pertenece a S?: {pertenece_3}")
```

### 1.3 Conjuntos Finitos e Infinitos
- ** Finito** Si tiene $n$ elementos distintos, donde $n$ es un numero entero no negativo.
- ** Infinito** En computacion, representamos conjuntos "Infinitos" mediante **generadores** o rangos, ya que la memoria es finita.

## 2.1 Subconjuntos
Se dice que $A$ es **subconjunto** de $B$ ($A \subseteq B$) si todo elemento de $A$ pertenece tambien a $B$ 
- **Subconjunto propio:** Si $A \subseteq B$ pero $A \neq B$.
- **Propiedad:** El conjunto vacio $\emptyset$ es subconjunto de todo conjunto.

### 2.2 Conjunto Potencia (Power Set)
Es el conjunto de todos los subconjuntos de un conjunto $S$, denotado como $\mathcal{P}(S)$.
Si $S$ tiene $n$ elementos, la cardinalidad de su conjunto potencia es $2^n$.


```python
# --- COMPARACION DE JERARQUIAS ---

A = {1, 2, 3}
B = {1, 2, 3, 4, 5}
C = {1, 2, 6}

# 1. Verificacion de Subconjuntos
es_sub_A_B = A.issubset(B)   # True: A esta contenido en B
es_sub_C_B = C.issubset(B)   # False: El 6 no esta en B

# 2. Operadores de comparacion (Sintaxis simplificada de Python)
# A <= B es equivalente a A.issubset(B)
def generar_conjunto_potencia(S):
    # Convertimos a lista para tener índices fijos
    elementos = list(S)
    n = len(elementos)
    potencia = []

    # El número total de subconjuntos DEBE ser 2 elevado a n
    num_subconjuntos = 2**n

    for i in range(num_subconjuntos):
        subconjunto = set()
        for j in range(n):
            # Verificamos si el bit j está encendido en el número i
            if (i & (1 << j)):
                subconjunto.add(elementos[j])
        potencia.append(subconjunto)

    return potencia

# --- PRUEBA DE FUEGO ---
S_test = {'a', 'b'}
resultado = generar_conjunto_potencia(S_test)

print(f"Conjunto S original: {S_test}")
print(f"Cardinalidad esperada: {2**len(S_test)}")
print(f"Resultado real: {resultado}")
print(f"Cardinalidad real: {len(resultado)}")
```
